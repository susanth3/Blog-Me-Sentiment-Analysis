### Function to flag keywords in title

```python
def keywordflag(keyword):
    keyword_flag = []
    length = len(df)
    for x in range(0,length):
        heading = df['title'][x]
        try:
            if keyword in heading:
                flag = 1
            else:
                flag = 0
        except:
            flag = 0
        keyword_flag.append(flag)
    return keyword_flag  
```

```python
key = keywordflag('murder')
```

```python
df['keyword_flag'] = pd.Series(key)
```

### Sentiment Analysis using VADER library

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
text = df['title'][16]
sentiment_analyser = SentimentIntensityAnalyzer()
sentiment_analyser.polarity_scores(text)
```

```python
length = len(df)
title_neg_sentiment = []
title_neu_sentiment = []
title_pos_sentiment = []
```

```python
for x in range(0,length):
    try:
        text = df['title'][x]
        sentiment_analyser = SentimentIntensityAnalyzer()
        results = sentiment_analyser.polarity_scores(text)
        neg = results['neg']
        pos = results['pos']
        neu = results['neu']
    except:
        neg = 0
        pos = 0
        neu = 0
    title_neg_sentiment.append(neg)
    title_neu_sentiment.append(neu)
    title_pos_sentiment.append(pos)
```

```python
df['title_neg_sentiment'] = pd.Series(title_neg_sentiment)
df['title_pos_sentiment'] = pd.Series(title_pos_sentiment)
df['title_neu_sentiment'] = pd.Series(title_neu_sentiment)
```
